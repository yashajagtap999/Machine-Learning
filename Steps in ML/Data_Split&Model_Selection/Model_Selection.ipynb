Preprocessing

```python
df_cleaned=df.copy()
```

```python
d2=df_cleaned["sex"].value_counts()
d2
```

```python
df_cleaned["sex"]=df_cleaned["sex"].map({"male":0,"female":1})
df_cleaned
```

```python
df_cleaned["smoker"].value_counts()
```

```python
#label encoding:--
df_cleaned["smoker"]=df_cleaned["smoker"].map({"no":0,"yes":1})
df_cleaned
```

```python
df_cleaned.rename(columns=
                  {
                      "sex":"is_male",
                      "smoker":"is_smoker"
                  })
```

```python
df_cleaned=pd.get_dummies(df_cleaned,columns=["region"],drop_first=True)
```

```python
df_cleaned
```

```python
df_cleaned=df_cleaned.astype(int)
df_cleaned
```

```python
df_cleaned["m_bmi"]=pd.cut(df_cleaned["bmi"],
                   bins=[10,20,30,40,float("inf")],
                   labels=["low","mid","high","vhigh"])
```

```python
df_cleaned=pd.get_dummies(df_cleaned,columns=["m_bmi"])
df_cleaned
```

```python
df_cleaned=df_cleaned.astype(int)
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
col=["age","bmi","children"]
scaler=StandardScaler()
df_cleaned[col]=scaler.fit_transform(df_cleaned[col])
```

```python
df_cleaned
```

Feature Extraction

```python
from scipy.stats import pearsonr # pearson correlation 
```

```python
selected_features=['age', 'sex', 'bmi', 'children', 'smoker', 'charges',
       'region_northwest', 'region_southeast', 'region_southwest', 'm_bmi_low',
       'm_bmi_mid', 'm_bmi_high', 'm_bmi_vhigh']
```

```python
correlations={
    feature:pearsonr(df_cleaned[feature],df_cleaned['charges'])[0]
    for feature in selected_features
}
```

```python
#converting correlations into dictionary
correlations_df=pd.DataFrame(list(correlations.items()),columns=["Feature","pearson correlations"])
correlations_df.sort_values(by=["pearson correlations"],ascending=True)
```

Feature selection

```python
final_df=df_cleaned[["smoker","charges","age"]]
final_df
```

```python
# Split the dataset :-- it is divided into 80% training and 20% testing
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=final_df.drop("charges",axis=1)
y=final_df["charges"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
#Model selection:--
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred_df=pd.DataFrame(y_pred)
y_pred_df
```
