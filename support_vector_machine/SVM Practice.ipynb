```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
label
```

```python
d1['survived']=label.fit_transform(d1['survived'])
```

```python
d1['sex']=label.fit_transform(d1['sex'])
```

```python
d1['age'].fillna(d1['age'].mean(),inplace=True)
d1
```

```python
d1=d1.drop(columns=['embarked', 'class', 'who', 'adult_male', 'deck', 'embark_town','alive', 'alone'])
```

```python
d1.isnull().sum()
```

```python
x=d1.drop('survived',axis=1)
y=d1["survived"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.9,random_state=42)
```

```python
from sklearn.svm import SVC
model=SVC(kernel='linear')
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
print(accuracy_score(y_pred,y_test))
```

```python
print(confusion_matrix(y_pred,y_test))
```
